```python
from matplotlib.patches import Circle, Rectangle, Arc

def draw_court(ax=None, color='black', lw=2):
    if ax is None:
        fig, ax = plt.subplots(figsize=(15, 7.5))  # scale to screen
    
    # Building the Court Floor
    floor = Rectangle((0, 0), 94, 50, linewidth=lw, color='tan', fill=True)
    ax.add_patch(floor)
    
    # out-of-bounds lines
    outer_lines = Rectangle((0, 0), 94, 50, linewidth=10, color='black', fill=False)
    ax.add_patch(outer_lines)

    # Center circle
    center_circle = Circle((94/2, 50/2), 6, linewidth=lw, color=color, fill=False)
    ax.add_patch(center_circle)
    
    # Half court line
    ax.plot([94/2, 94/2], [0, 50], color=color, linewidth=lw)
    
# Building the Three Point Lines:

    # 3-point arc (left side)
    three_point_arc_left = Arc((4.7, 25), 23.9*2, 23.9*2, theta1=-67, theta2=67, linewidth=lw, color='black')
    ax.add_patch(three_point_arc_left)

    # Three point line upper left straight side part
    ax.plot([0, 14], [47, 47], color='black', linewidth=lw)
    
    # Three point line lower left straight side part
    ax.plot([0, 14], [3, 3], color='black', linewidth=lw)

    # 3-point arc (right side)
    three_point_arc_left = Arc(((94-4.7), 25), 23.9*2, 23.9*2, theta1=113, theta2=247, linewidth=lw, color='black')
    ax.add_patch(three_point_arc_left)

    # Three point line upper right straight side part
    ax.plot([80, 94], [47, 47], color='black', linewidth=lw)
    
    # Three point line lower right straight side part
    ax.plot([80, 94], [3, 3], color='black', linewidth=lw)

# Building the Key and Free throw circle:

    # Left side Key
    LKey = Rectangle((0, 19), 19, 12, linewidth=lw, color='black', fill=False)
    ax.add_patch(LKey)

    #Left side Key free throw circle, outside the Key
    key_arc_left_up = Arc(((19), 25), 12, 12, theta1=-90, theta2=90, linewidth=lw, color='black')
    ax.add_patch(key_arc_left_up)

    #Left side free throw circle, in the Key
    key_arc_left_down = Arc((19, 25), 12, 12, theta1=90, theta2=270, linewidth=lw, color='black', linestyle='dashed')
    ax.add_patch(key_arc_left_down)

    #Left side Outer Key
    LOuterKey = Rectangle((0, 17), 19, 16, linewidth =lw, color='black', fill=False)
    ax.add_patch(LOuterKey)

    #Left side Charge Circle
    charge_circle_left = Arc((5.25, 25), 6, 6, theta1=-90, theta2=90, linewidth=lw, color='black')
    ax.add_patch(charge_circle_left)
    #Left side Charge Circle Straight parts
    ax.plot([4, 5.25], [28, 28], color='black', linewidth=lw)




    #Right side Key
    RKey = Rectangle((94-19, 19), 19, 12, linewidth=lw, color='black', fill=False)
    ax.add_patch(RKey)

    #Right side Key free throw circle, outside the Key
    key_arc_right_up = Arc(((94-19), 25), 12, 12, theta1=90, theta2=270, linewidth=lw, color='black')
    ax.add_patch(key_arc_right_up)

    #Right side free throw circle, in the Key
    key_arc_right_down = Arc((94-19, 25), 12, 12, theta1=-90, theta2=90, linewidth=lw, color='black', linestyle='dashed')
    ax.add_patch(key_arc_right_down)

    #Right side Outer Key
    ROuterKey = Rectangle((94-19, 17), 19, 16, linewidth =lw, color='black', fill=False)
    ax.add_patch(ROuterKey)





#Making the Court fit the screen
    # Set limits and aspect
    ax.set_xlim(0, 94)
    ax.set_ylim(0, 50)
    ax.set_aspect('equal', adjustable='box')
    ax.axis('off')  # hides ticks
    
    
    return ax

draw_court()
plt.show()
```
